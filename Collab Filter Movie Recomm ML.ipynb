```python
df_movies=pd.read_csv('D:/IBM AI SGUnited/DATA SCIENCE/Project/Data/2018/V22movies.csv')
df_ratings=pd.read_csv('D:/IBM AI SGUnited/DATA SCIENCE/Project/Data/2018/V22ratings.csv')

#df_movies=pd.read_csv(movies, usecols=['movie_id','title'])
#df_ratings=pd.read_csv(ratings, usecols=['user_id','movie_id','rating']
```

```python
movies_users=df_ratings.pivot(index='movie_index', columns='user_index',values='rating')
movies_users
```

```python
movies_users=df_ratings.pivot(index='movie_index', columns='user_index',values='rating').fillna(0)
movies_users
```

```python
movie_user_matrix=csr_matrix(movies_users.values)
```

next we build a model using the nearest neighbors function, KNN(known nearest neighbor)

```python
#model_KNN = NearestNeighbors(metric='p_norm',algorithm ='kd_tree', n_neighbors=20)
#model_KNN = NearestNeighbors(metric='cosine',algorithm ='brute', n_neighbors=20)
#apply the unsupervised learning algorithm, 

model_KNN = NearestNeighbors(n_neighbors = 20, algorithm = 'ball_tree')
```

next we train our model, by simple using the 'fit' method

```python
model_KNN.fit(movie_user_matrix)
```

```python
def recommender(movie_name, data,model, n_recommendations ):
    model.fit(data)
    idx=process.extractOne(movie_name, df_movies['title'])[2]
    print('Movie Selected: ',df_movies['title'][idx], 'Index: ',idx)
    print('Searching for recommendations.....')
    distances, indices=model.kneighbors(data[idx], n_neighbors=n_recommendations)
    for i in indices:
        print(df_movies['title'][i].where(i!=idx))
    
recommender('Bruno', movie_user_matrix, model_KNN,20)
```
